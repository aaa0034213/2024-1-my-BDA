## pandas 함수 조합하기

```python
# suv 추출
# 합산 연비 변수 만들기
# 제조 회사별로 분리
# 합산 연비 평균 구하기
# 내림차순 정렬
# 1~5위까지 출력, 시험 예상

mpg.query('category == "suv"') \
   .assign(total = (mpg['hwy'])+(mpg['cty'])/2) \
   .groupby('manufacturer') \
   .agg(mean_tot= ('total','mean')) \
   .sort_values('mean_tot', ascending=False)\
   .head()
```

## 세로로 합치기: concat()

```python
# 학생 1~5번 시험 데이터 만들기
group_a = pd.DataFrame({'id'   : [1, 2, 3, 4, 5],
                        'test' : [60, 80, 70, 90, 85]})

# 학생 6~10번 시험 데이터 만들기
group_b = pd.DataFrame({'id'   : [6, 7, 8, 9, 10],
                        'test' : [70, 83, 65, 95, 80]})
```

```python
group_a
```

```python
group_b
```

```python
group_all = pd.concat([group_a,group_b])
group_all
```

```python
group_all.reset_index()         # 원 index가 열로 이동, 새로운 시퀀스 지정
```

```python
group_all.reset_index(drop = True)    # 열 index 제거, 새로운 시퀀스 지정
```

### 심화 학습: [공식 매뉴얼 문서](https://pandas.pydata.org/docs/reference/api/pandas.merge.html#pandas.merge)

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
df1
```

```python
df2
```

```python
# key 및 rkey 열에 df1 및 df2를 병합, 공통 키가 다른 경우
pd.merge(df1, df2, left_on = 'lkey', right_on = 'rkey')
```

```python
df1.merge(df2, left_on='lkey', right_on='rkey', suffixes=('_left', '_right'))
```
